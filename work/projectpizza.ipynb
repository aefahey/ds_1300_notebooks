```python
import folium
loc_center = [df['location_lat'].mean(), df['location_lng'].mean()]
```

```python
map1 = folium.Map(location = loc_center, tiles='Openstreetmap', zoom_start = 5, control_scale=True)
for index, loc in df.iterrows():
    folium.CircleMarker([loc['location_lat'], loc['location_lng']],     radius=2, weight=5, popup=loc['Name']).add_to(map1)
folium.LayerControl().add_to(map1)
map1
```
