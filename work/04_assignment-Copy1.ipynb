```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
data = pd.read_csv('../data/state_facts.tsv', delimiter='\t')
```

```python
plt.figure
data.groupby("State") ["Capital_pop"].sum().plot(kind='bar', figsize=(20,10), legend=True, color='purple')
plt.title('Capital Population Of Each State')
plt.xlabel('State')
plt.ylabel('Capital Population')
plt.show ()
```

```python
#5: This plot shows the each state in the U.S. (x-axis) & their corresponding captial population (y-axis) through a bar chart.
```

```python
#6: I choose to take these two points of data because capital population size doesn't always correlate with the most popular state. Such as Ohio, which I wouldn't of thought has close to the same capital size as Texas. Furthermore, I choose to make this plot a bar chart because it is fairly simple to look at and see overall what states have the largest capital population and also makes it easy to compare among states.
```
